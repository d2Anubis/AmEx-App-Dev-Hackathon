### Tweet preprocessing
1. removing the @mentions and links from tweets to keep only the main tweet by the user
2. using sentiment analysis tool by NLTK to get the sentiment of the tweet as Positive/ Negative/ Neutral
3. assigning sentiment and sentiment scores to the above pre-processed tweets
4. removing stop words from processed tweets to create a dictionary later

```
#preprocessing tweets
#removing @mention and urls
import re
def remove_mention_n_links(tweet):
    return(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split()))

df1['raw_tweet'] = df1['tweet'].apply(remove_mention_n_links)
```

```
#downloading nltk libraries and functions that we need
import nltk
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('punkt')
```

```
#assigning sentiments and sentiment score to tweets
#we are using a pre-defined model for sentiment analysis. 
#We can also use spacy and other deep learning methods for sentiment analysis if we have labeled dataset

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiment_analysis = SentimentIntensityAnalyzer()

def get_sentiment(tweet):
    score = sentiment_analysis.polarity_scores(tweet)['compound']
    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"
    
def get_sentiment_score(tweet):
    return sentiment_analysis.polarity_scores(tweet)['compound']

#calling the sentiment analysis function to assign sentiment to each tweet
df1['sentiment'] = df1['raw_tweet'].apply(get_sentiment)
df1['sentiment_score'] = df1['raw_tweet'].apply(get_sentiment_score)
```

```
# removing stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))

def remove_stop_words(tweet):
    tweet = word_tokenize(tweet)
    tweet = [w for w in tweet if not w in stop_words]
    return " ".join(tweet)

#calling the function to remove the stopwords
df1['tweet_wo_stop_words'] = df1['raw_tweet'].apply(remove_stop_words)
```

### Creating 3 different dictionaries. 

1. Complete dictionary containing all the words
2. Dictionary for positive words
3. Dictionary of negative words

Note : We are only considering customer tweets for analysis

```
#removing some of the most occured words
other_most_occured_words = ['card', 'i','hi','please','amex','u','s','number','us','personal','release','back',
                            'reaching','we','thanks','which','thank','would','express','american','identifiable',
                            'pls','sorry','get','help','concerning','need','customer','hello','good','new','use',
                            'b','like','amp','based','rachel','morning','meg','know','if','thx','access','it',
                            'the','able','one','is','serve','members','you','time','w','may','want','contact','read',
                            'see','k','online','twitter','using','better','directly','what','can','team','d','look',
                            'via','bluebird','clarissa','not','offer','let','tweeting','do','make','happy','day',
                            'gift','concern','fee','received','phone','so', 'got', 'still', 'my', 'how', 'trying', 'since',
                            'know', 'even', 'told' ]

dict_all = " ".join(df1[df1.username != "askamex"].loc[:, "tweet_wo_stop_words"].values)
dict_pos = " ".join(df1[(df1.sentiment == "Positive") & (df1.username != "askamex")].loc[:, "tweet_wo_stop_words"].values)
dict_neg = " ".join(df1[(df1.sentiment == "Negative") & (df1.username != "askamex")].loc[:, "tweet_wo_stop_words"].values)

print("Corpus created.")
all_dict = {}
pos_dict = {}
neg_dict = {}

#creating frequency dictionary for each word
for word in word_tokenize(dict_all):
    if word.lower() not in other_most_occured_words:
        if word.lower() in all_dict:
            all_dict[word.lower()] += 1
        else:
            all_dict[word.lower()] = 1

print("Overall dictionary created.")
        
for word in word_tokenize(dict_pos):
    if word.lower() not in other_most_occured_words:
        if word.lower() in pos_dict:
            pos_dict[word.lower()] += 1
        else:
            pos_dict[word.lower()] = 1

print("Positive dictionary created.")
        
for word in word_tokenize(dict_neg):
    if word.lower() not in other_most_occured_words:
        if word.lower() in neg_dict:
            neg_dict[word.lower()] += 1
        else:
            neg_dict[word.lower()] = 1

print("Negative dictionary created.")

dict_all = pandas.DataFrame(sorted(all_dict.items(), key=lambda x: x[1], reverse=True)[:50], columns=['word', 'count'])
dict_pos = pandas.DataFrame(sorted(pos_dict.items(), key=lambda x: x[1], reverse=True)[:40], columns=['word', 'count'])
dict_neg = pandas.DataFrame(sorted(neg_dict.items(), key=lambda x: x[1], reverse=True)[:40], columns=['word', 'count'])
```

```
dict_all.to_csv("analysis_data2/dictionary_all_consumer.csv", index=False)
dict_pos.to_csv("analysis_data2/dictionary_positive_words_consumer.csv", index=False)
dict_neg.to_csv("analysis_data2/dictionary_negative_words_consumer.csv", index=False)
```
